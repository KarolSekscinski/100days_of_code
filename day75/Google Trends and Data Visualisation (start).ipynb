```python
def find(string, pattern):
    string = string.lower()
    pattern = pattern.lower()
    n = len(pattern)
    
    count = []
    # find initial hashes for i = 0
    hash_p = sum(map(lambda x: ord(x), list(pattern)))
    hash_s = sum(map(lambda x: ord(x), list(string[:n])))
    
    for i in range(len(string)-n+1):
        if hash_p == hash_s and string[i:i+n] == pattern:
            count.append(i)
            
        # update hash for next character
        if i+n < len(string):
            hash_s = hash_s - ord(string[i]) + ord(string[i+n])
        
    return len(count)
            
    
print(find("ala Ma kota kota kota", "kota"))
print(find("ala Ma kota", "ma"))
print(find("ala Ma kota", " k"))
print(find("ala Ma kota", "ss"))
```
